# Dataset preparation

Unlike other solutions, the dataset for SSD needs to be modified manually if exported from roboflow in Pascal VOC format.
The following steps are necessary:
- Images need to be exported at 300x300 pixel resolution
- Assuming a train-test-val split is present, move the contents of "valid" to "train" and delete folder "valid"
- create new folder "Annotations"
- move all .xml files from "test" and "train" to "Annotations"
- navigate to folder "train" and create a text file containing the name of every image file, name the .txt file "trainval.txt"
- repeat for the folder "test", name the .txt file "test.txt"
- create new folder "JPEGImages"
- move the remaining contents (images) from "test" and "train" to "JPEGImages"
- create new folder "ImageSets"
- create new folder "Main" inside "ImageSets"
- move "trainval.txt" and "test.txt" to folder "Main"
- create text file "labels.txt" in main directory containing every class label, one per line


# Training

Different NN architectures are provided:
- vgg16-ssd
- mb1-ssd
- mb1-ssd-lite
- sq-ssd-lite
- mb2-ssd-lite
- mb3-large-ssd-lite
- mb3-small-ssd-lite

Parameters to change for a different architecture:
- "--net" --> change to one of the listed above
- "--base_net" --> provide base net if possible, else delete parameter
- optional: parameter tuning for better results
