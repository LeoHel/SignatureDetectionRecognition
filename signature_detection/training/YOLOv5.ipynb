# Modify model configuration

In order for the model to train properly, the configuration has to be updated according to the number of desired classes.

```
import yaml
import re

with open("/content/data.yaml", 'r') as stream:
    num_classes = str(yaml.safe_load(stream)['nc'])

with open("/content/yolov5/models/yolov5s.yaml", 'r') as y:
  cust = y.read()

with open('/content/yolov5/models/custom_yolov5s.yaml', 'w') as c:
  cust = re.sub('nc: 80', 'nc: {}'.format(num_classes), cust)
  c.write(cust)
```
