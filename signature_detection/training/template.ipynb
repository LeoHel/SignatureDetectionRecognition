# Create data set vector file

Using a single signature cutout ("test.png"), a vector file is created.
500 examples containing a single signature of width 15 and height 30 with background from the racoon dataset are created.

```python
!opencv_createsamples -num 500 -vec vector.v -img test.png -w 15 -h 30 -bg neg.txt
```

# Training

The cascade is trained. Important: 
- width and height must match the vector file examples
- numPos must be smaller than the number created in the previous step
- numNeg must not exceed the number of available background images
- vec is the parameter for the vector file
- bg can use the same txt file as example creation

```python
!opencv_traincascade -data /content/data -vec vector.v -bg neg.txt -numPos 450 -numNeg 200 -w 15 -h 30
```
